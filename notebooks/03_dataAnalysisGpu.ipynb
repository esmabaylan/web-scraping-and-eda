```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("/work/data/processed/02_extract_os_features.csv")
df=dataset.copy()
df.dropna(subset=["cpu"], inplace=True)
```

```python
df.loc[df['gpu'].str.contains(r'(?:rtx|gtx|mx)\s*0\d+', regex=True, na=False), 'gpu'] = np.nan
df['gpu'] = df['gpu'].str.replace(
    r'rtx\s*(?:pro|a)?\s*(500|1000|2000|3000|3500)\s*(?:ada|a)?', 
    r'rtx a\1', 
    regex=True
)
```

```python
gpu_mapping = {
    'intel uhd graphics': 1, 'amd radeon': 1, 'radeon 530': 1,
    
    'intel iris xe': 2, 'intel arc graphics': 2, 'm1': 2, 'm2': 2,
    'mx330': 2, 'mx350': 2, 'mx550': 2, 'mx570': 2, 'mx570a': 2,
    'radeon 680m': 2,
    
    'radeon 780m': 3, 'radeon 890m': 3, 'm3': 3,
    'rtx 2050': 3, 'rtx a500': 3, 'rtx a1000': 3,
    
    'rtx 3050': 4, 'rtx 3050ti': 4, 'rtx 4050': 4, 'rtx 5050': 4, 
    'rtx a2000': 4,
    
    'rtx 3060': 5, 'rtx 4060': 5, 'rtx 5060': 5, 
    'rtx a3000': 5, 'rtx a3500': 5,
    
    'rtx 3070': 6, 'rtx 4070': 6, 'rtx 5070': 6, 'rtx 5070ti': 6,
    
    'rtx 5080': 7, 'rtx 5090': 7
}

df['gpu_priority'] = df['gpu'].map(gpu_mapping)
```

```python
df.to_csv("/work/data/processed/03_extract_gpu_features.csv", index=False)
```
