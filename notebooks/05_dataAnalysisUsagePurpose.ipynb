```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("/work/data/processed/04_extract_memory_features.csv")
df=dataset.copy()
```

```python
def determine_usage_purpose(row):
    cpu = str(row['cpu']).lower()
    gpu = str(row['gpu']).lower()
    

    try:
        ram_val = int(str(row['ram']).lower().replace('gb', '').strip())
    except:
        ram_val = 8 

    if any(keyword in gpu for keyword in ['rtx', 'gtx', 'rx ', 'radeon rx', 'rtx4050']) or \
       any(keyword in cpu for keyword in [' h', 'hs', 'hx']):
        return 'Gaming / Design'
    

    elif any(keyword in cpu for keyword in ['celeron', 'pentium', 'athlon', 'core 3', 'i3', 'ryzen 3']) or (ram_val <= 8):
        return 'Daily'
    

    else:
        return 'Office'
```

```python
df['usage_purpose'] = df.apply(determine_usage_purpose, axis=1)

print(df[['brand', 'name', 'cpu', 'gpu', 'ram', 'usage_purpose']].head(10))

df.to_csv('/work/data/processed/05_extract_usage_purpose_features.csv', index=False)
```
