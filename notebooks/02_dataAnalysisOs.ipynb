```python
import pandas as pd
```

```python
dataframe = pd.read_csv("/work/data/processed/01_extract_cpu_features.csv")
df=dataframe.copy()
```

```python
df["os"]=df["os"].str.replace(r"(Windows 10$)", "Windows 10 Home", regex=True)
df["os"]=df["os"].str.replace(r"(Windows 11$)", "Windows 11 Home", regex=True)
```

```python
df["os"].unique()
```

```python
os_mapping = {
    'FreeDOS': 1,
    'Linux': 1,
    'Android / HarmonyOS': 1,
    'Windows': 2,
    'Windows 10 Home': 2,
    'Windows 10 Pro': 3,
    'Windows 11 Home': 3,
    'Windows 11 Pro': 4,
    'macOS': 4,
}

df['os_priority'] = df['os'].map(os_mapping)
```

```python
df.to_csv("/work/data/processed/02_extract_os_features.csv", index=False)
```
